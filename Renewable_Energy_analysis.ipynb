***Highlight Global Trends in Renewable Energy***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
file_path = '/content/sample_data/IRENA_Stats_Extract_ 2024_H1_V1.xlsx'
raw_data = pd.read_excel(file_path, sheet_name='All Data', engine='openpyxl')
```

```python
print("First few rows of the raw dataset:")
raw_data.head()
```

```python
print("Raw data shape:", raw_data.shape)
```

```python
print("Column names in the raw dataset:")
print(raw_data.columns)
```

```python
print("Sample of raw data (including first 10 rows):")
print(raw_data.head(10))
```

```python
# Print unique technologies
print("Unique technologies in the dataset:")
print(raw_data['Technology'].unique())
```

```python
# Step 1: Data Cleaning

# 1. Retain relevant columns
columns_to_keep = [
    'Year', 'Region', 'Country', 'Technology', 'Electricity Installed Capacity (MW)'
]
data = raw_data[columns_to_keep]
```

```python
# Print all column names to confirm
print("Column names in the dataset:")
print(data.columns)
```

```python
# Display the first few rows
print("First few rows of the dataset:")
print(data.head())
```

```python
# Print the unique technologies in the Technology column
print("Unique technologies in the dataset:")
print(data['Technology'].unique())
```

```python
# 2. Filter rows related to Renewable Energy
renewable_technologies = [
    "Geothermal energy", "Renewable hydropower", "Marine energy",
    "Solar energy", "Solar photovoltaic", "Solar thermal energy",
    "Onshore wind energy", "Offshore wind energy", "Solid biofuels",
    "Biogas", "Liquid biofuels", "Renewable municipal waste", "Other renewable energy"
]
data = data[data['Technology'].isin(renewable_technologies)]
```

```python
# 3. Handle missing values
# Drop rows where 'Electricity Installed Capacity (MW)' or 'Year' is missing
data = data.dropna(subset=['Electricity Installed Capacity (MW)', 'Year'])
```

```python
# Ensure 'Year' is numeric for sorting
data['Year'] = data['Year'].astype(int)
```

```python
# Step 2: Objective 1 - Global Trends Analysis

# 1. Aggregate data by Year and Technology
trend_data = data.groupby(['Year', 'Technology'])['Electricity Installed Capacity (MW)'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
for tech in renewable_technologies:
    tech_data = trend_data[trend_data['Technology'] == tech]
    plt.plot(tech_data['Year'], tech_data['Electricity Installed Capacity (MW)'], label=tech)

plt.title('Global Trends in Renewable Energy Installed Capacity (MW)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Installed Capacity (MW)', fontsize=12)
plt.legend(title='Technology', fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

**Plotting Top 5 Renewable Energies**

```python
# Aggregate total installed capacity by Technology
total_capacity = trend_data.groupby('Technology')['Electricity Installed Capacity (MW)'].sum().reset_index()

# Sort by total capacity and get the top 5
sorted_capacity = total_capacity.sort_values(by='Electricity Installed Capacity (MW)', ascending=False).head(5)
```

```python
# Filter trend data for top 5 technologies
top_5_technologies = sorted_capacity['Technology'].tolist()
filtered_trend_data = trend_data[trend_data['Technology'].isin(top_5_technologies)]
```

```python
plt.figure(figsize=(12, 6))
for tech in top_5_technologies:
    tech_data = filtered_trend_data[filtered_trend_data['Technology'] == tech]
    plt.plot(tech_data['Year'], tech_data['Electricity Installed Capacity (MW)'], label=tech)

plt.title('Trends in Installed Capacity for Top 5 Renewable Energy Technologies', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Installed Capacity (MW)', fontsize=12)
plt.legend(title='Technology', fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

**Performing Correlation Analysis with GDP and Population**

```python
# Load GDP and Population data
gdp_data = pd.read_excel('/content/sample_data/GDP.xlsx', sheet_name='Data')
population_data = pd.read_excel('/content/sample_data/population.xlsx', sheet_name='Data')
```

```python
# Preview the datasets
print("GDP Data Sample:")
print(gdp_data.columns)

print("Population Data Columns:")
print(population_data.columns)
```

```python
# Reshape GDP data
gdp_long = gdp_data.melt(
    id_vars=['Country Name', 'Country Code'],
    var_name='Year',
    value_name='GDP'
)

# Reshape Population data
population_long = population_data.melt(
    id_vars=['Country Name', 'Country Code'],
    var_name='Year',
    value_name='Population'
)
```

```python
# Convert 'Year' to numeric for easier merging
gdp_long['Year'] = pd.to_numeric(gdp_long['Year'], errors='coerce')
population_long['Year'] = pd.to_numeric(population_long['Year'], errors='coerce')

# Print samples of reshaped data
print("Reshaped GDP Data Sample:")
print(gdp_long.head())

print("Reshaped Population Data Sample:")
print(population_long.head())
```

```python
# Drop missing values
gdp_long.dropna(subset=['GDP'], inplace=True)
population_long.dropna(subset=['Population'], inplace=True)
```

```python
# Merge GDP and Population data
gdp_population_data = pd.merge(gdp_long, population_long, on=['Country Name', 'Country Code', 'Year'], how='inner')
```

```python
# Preview the merged dataset
print("Merged GDP and Population Data Sample:")
print(gdp_population_data.head())
```

```python
gdp_population_data.rename(columns={'Country Name': 'Country'}, inplace=True)
```

```python
# Merge with renewable energy data
final_data = pd.merge(data, gdp_population_data, on=['Country', 'Year'], how='inner')

# Preview the final dataset
print("Final Dataset Sample:")
print(final_data.head())
```

```python
# Save the cleaned data to an Excel file
output_file_path = 'Cleaned_Renewable_GDP_Population_Data.xlsx'
final_data.to_excel(output_file_path, index=False, engine='openpyxl')
```

```python
# Correlation Analysis
correlation_matrix = final_data[['Electricity Installed Capacity (MW)', 'GDP', 'Population']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

***Compare Regional Performance in Renewable Energy Adoption***

```python
# Aggregate data by Region and Year
regional_data = data.groupby(['Year', 'Region'])['Electricity Installed Capacity (MW)'].sum().reset_index()
```

```python
# Step 3: Visualization - Compare Regional Performance Over Time

plt.figure(figsize=(12, 6))
regions = regional_data['Region'].unique()

for region in regions:
    region_data = regional_data[regional_data['Region'] == region]
    plt.plot(region_data['Year'], region_data['Electricity Installed Capacity (MW)'], label=region)

plt.title('Regional Performance in Renewable Energy Installed Capacity (MW)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Installed Capacity (MW)', fontsize=12)
plt.legend(title='Region', fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Step 4: Highlighting Top Performing Region
# Find the total installed capacity for each region
total_capacity_by_region = data.groupby('Region')['Electricity Installed Capacity (MW)'].sum().reset_index()
```

```python
# Sort by total capacity
top_regions = total_capacity_by_region.sort_values(by='Electricity Installed Capacity (MW)', ascending=False)
```

```python
# Bar Plot for Total Installed Capacity by Region
plt.figure(figsize=(10, 6))
plt.bar(top_regions['Region'], top_regions['Electricity Installed Capacity (MW)'], color='deepskyblue')
plt.title('Total Renewable Energy Installed Capacity by Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Installed Capacity (MW)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Regional Growth Rates**
Calculate the annual growth rates of renewable energy capacity for each region.

```python
# Calculate annual growth rates for each region
regional_data['Growth Rate (%)'] = regional_data.groupby('Region')['Electricity Installed Capacity (MW)'].pct_change() * 100

# Display average growth rate for each region
avg_growth_by_region = regional_data.groupby('Region')['Growth Rate (%)'].mean().reset_index()
avg_growth_by_region = avg_growth_by_region.sort_values(by='Growth Rate (%)', ascending=False)

print("Average Growth Rate by Region (%):")
print(avg_growth_by_region)
```

**Regional Contributions to Global Capacity**
Show the percentage contribution of each region to global renewable energy capacity.

```python
# Calculate global total installed capacity
global_total_capacity = data['Electricity Installed Capacity (MW)'].sum()

# Calculate percentage contribution of each region
region_contributions = total_capacity_by_region.copy()
region_contributions['Percentage Contribution (%)'] = (region_contributions['Electricity Installed Capacity (MW)'] / global_total_capacity) * 100

print("Regional Contributions to Global Capacity:")
print(region_contributions)
```

**Heatmap of Regional Performance**
 Visualizes the total capacity of different technologies for each region using a heatmap.

```python
tech_by_region = data.groupby(['Region', 'Technology'])['Electricity Installed Capacity (MW)'].sum().reset_index()
```

```python
# Corrected pivot table code
heatmap_data = tech_by_region.pivot(index='Region', columns='Technology', values='Electricity Installed Capacity (MW)').fillna(0)
```

```python
# Step 3: Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlGnBu', cbar=True)
plt.title('Renewable Energy Capacity by Region and Technology (MW)', fontsize=14)
plt.xlabel('Technology', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.tight_layout()
plt.show()
```
